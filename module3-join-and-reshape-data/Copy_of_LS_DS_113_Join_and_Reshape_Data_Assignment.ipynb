```python
import pandas as pd
import numpy as np

table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T

table2
```

```python
# table2 -> tidy
table2
```

```python
table2.reset_index(inplace=True)
```

```python
table2
```

```python
tidy = table2.melt(id_vars='index',
                   value_vars=['John Smith', 'Jane Doe', 'Mary Johnson'])
```

```python
tidy.rename(columns={'index':'trt',
                     'variable':'names',
                     'value':'results'},
            inplace=True)
```

```python
tidy['trt'] = tidy['trt'].str.replace('treatment','')
```

```python
tidy
```

```python
# tidy -> table2
wide = tidy.pivot_table(index='trt',columns='names',values='results')
```
